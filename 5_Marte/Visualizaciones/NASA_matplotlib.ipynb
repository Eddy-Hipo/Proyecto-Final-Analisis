```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
```

```python
conn=sqlite3.connect("mars.db")
dfmars=pd.read_sql_query("SELECT * FROM NASA_posts", conn)
# dfmars=dfmars.set_index('index')
```

```python
ax1 = dfmars.plot(y='Likes', color = '#405de6')
ax2 = dfmars.plot(y='Comments',color = '#c13584', ax=ax1)
plt.title('Valoracin de las publicaciones sobre Marte (2021)')
plt.legend(['Me gustas','Comentarios'])
ax2.set_xlabel('ID de la publicacin')
ax1.set_ylabel('Cantidad alcanzada')
plt.grid(color='#8e8e93', linestyle='dotted', linewidth=0.5)
plt.savefig('NASA1.svg', format='svg', dpi=1200)
plt.show()
```

```python
max_Likes = dfmars['Likes'].max()
max_Comments = dfmars['Comments'].max()
max_Shares = dfmars['Shares'].max()
num = [max_Likes, max_Comments, max_Shares]
title=["Me gustas","Comentarios","Compartidos"]
plt.bar(title, num, color = ['#833ab4','#fd1d1d','#fcaf45'], width = 0.8)
plt.title('Valores mximos alcanzados por noticias de Marte (2021)')
plt.grid(color='#8e8e93', linestyle='dotted', linewidth=0.5)
plt.savefig('NASA2.svg', format='svg', dpi=1200)
plt.show()
```

```python
mean_Likes = dfmars['Likes'].mean()
mean_Comments = dfmars['Comments'].mean()
mean_Shares = dfmars['Shares'].mean()
num = [mean_Likes, mean_Comments, mean_Shares]
titles = ["Me gustas","Comentarios","Compartidos"]
colores = ["#5851db","#f56040","#c13584"]
explode = (0.1, 0, 0)
plt.pie(num, labels=titles, autopct="%0.2f %%", colors=colores, shadow=True, explode=explode)
plt.axis("equal")
plt.title('Promedio de valoraciones (2021)')
plt.savefig('NASA3.svg', format='svg', dpi=1200)
plt.show()
```

```python
important_post=dfmars[(dfmars['Likes']>70000)]
n1=int(important_post['Likes'])
n2=int(important_post['Comments'])
n3=int(important_post['Shares'])
x_values = [n1, n2, n3]
y_values = ['Me gustas','Comentarios','Compartidos']
plt.barh(y_values, x_values, align='center',tick_label=y_values, color = ['#833ab4','#fd1d1d','#fcaf45'])
plt.xlabel('Cantidad alcanzada')
plt.title('Valoraciones de evento importante sobre Marte (19/02/2021)')
plt.grid(color='#8e8e93', linestyle='dotted', linewidth=0.5)
plt.savefig('NASA4.svg', format='svg', dpi=1200)
plt.show()
```

```python
shared=dfmars[(dfmars['Shares']!=0)]
y = shared['Shares']
x = shared['index']
plt.plot(x, y, "-o", color ='#fcaf45')
plt.grid(color='#8e8e93', linestyle='dotted', linewidth=0.5)
plt.title('Publicaciones compartidas de Marte (2021)')
plt.xlabel('ID de la publicacin')
plt.ylabel('Cantidad alcanzada')
plt.savefig('NASA5.svg', format='svg', dpi=1200)
plt.show()
```
